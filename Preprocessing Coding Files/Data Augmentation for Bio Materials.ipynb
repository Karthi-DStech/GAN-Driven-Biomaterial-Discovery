```python
from PIL import Image
import os

def rotate_and_save_all_images(source_dir, dest_dir):

    os.makedirs(dest_dir, exist_ok=True)


    for filename in os.listdir(source_dir):
        if filename.endswith('.png'):  # Check for PNG files

            file_path = os.path.join(source_dir, filename)
            

            original_image = Image.open(file_path)
            

            original_image.save(os.path.join(dest_dir, filename))
            

            for angle in [90, 180, 270]:

                rotated_image = original_image.rotate(angle, expand=True)
                

                new_filename = f"{os.path.splitext(filename)[0]}_{angle}.png"
                

                rotated_image.save(os.path.join(dest_dir, new_filename))


source_dir = '/Users/karthik/GANS/Original-Bio-Material-Data'

# Specify the destination directory where you want to save the images
dest_dir = '/Users/karthik/4by4-Rotated-Bio-Material-Data'

# Perform the rotation and save for all images
rotate_and_save_all_images(source_dir, dest_dir)
```

```python
import glob
import cv2
from PIL import Image
import os
```

```python
input_path_gr = '/Users/karthik/2by2-Rotated-Bio-Material-Data/'
output_path_gr = '/Users/karthik/2by2-Resized_Rotated-Topo'
```

```python
if not os.path.exists(output_path_gr):
    os.makedirs(output_path_gr)
```

```python
# Iterate over all the image files in the input directory
for image_file in glob.glob(os.path.join(input_path_gr, '*.png')):
    # Assuming PNG format, change accordingly if different
    # Read the image
    img = cv2.imread(image_file)
    
    # Resize the image
    resized_img = cv2.resize(img, (256, 256), interpolation=cv2.INTER_AREA)
    
    # Construct the output file path and save the image
    base_name = os.path.basename(image_file)
    output_file_path = os.path.join(output_path_gr, base_name)
    cv2.imwrite(output_file_path, resized_img)

print("Resizing complete!")
```
